````python
import pandas as pd

file1 = 'states/states.tsv'
file2 = 'sensors/Sensors_OperativeTemperature_Hourly.d6o'
````

Read the tables in a dataframe. 
For the second dataframe some additional data is needed:

 - ```sep```: What is used to separate different entries, tabs in this case
 - ```header```: Indiciate we have no header for our columns in the file
 - ```names```: Provide the column names
 - ```index_col```: Do not use one of the columns as the index


````python
data1 = pd.read_table('states/states.tsv')
data2 = pd.read_csv('sensors/Sensors_OperativeTemperature_Hourly.d6o', sep='\t', skiprows=15, header=None, names=['Time [d]', 'Sensor 4000', 'Sensor 4001'], index_col=False)
````

Merge the two dataframes based on the sorted ```'Time [d]'``` column.

````python
data = pd.merge(data1, data2, on='Time [d]', how='outer', sort=True)
data
````

Test if the data in the final df is the same as in the original

````python
difference = 0
for i in data2['Time [d]']:
    #Get the old data for time i
    old = data2.ix[data2[data2['Time [d]']==i].index[0]]['Sensor 4000']
    #Get the new data for time i
    new = data.ix[data[data['Time [d]']==i].index[0]]['Sensor 4000']
    #Calculate the difference
    difference += abs(old-new)
````
